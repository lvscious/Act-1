```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data
df = pd.read_csv('Walmart_Sales_300_rows.csv')
sales = df['Weekly_Sales']

# Set up the plot
plt.figure(figsize=(12, 7))
n_bins = 25
counts, bins, patches = plt.hist(sales, bins=n_bins, edgecolor='white', alpha=0.8)

patches[0].set_facecolor('#e74c3c') # Red for the lowest sales bin
blue_color = '#008080's
for i in range(1, len(patches)):
    patches[i].set_facecolor(blue_color)


plt.title('Distribution of Weekly Sales: ', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Weekly Sales (in Millions of Dollars)', fontsize=13, labelpad=10)
plt.ylabel('Number of Weeks', fontsize=13, labelpad=10)


# Report the numbers again for the text
lowest_val = sales.min()
highest_val = sales.max()
print(f"Min: {lowest_val}, Max: {highest_val}")
```
