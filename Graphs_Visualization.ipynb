```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load your dataset
df = pd.read_excel('/content/E-Commerse Dataset.xlsx')

# 2. Prepare the data
df['Date'] = pd.to_datetime(df['Date'])

# Bar Chart
plt.figure(figsize=(10, 6))
category_sales = df.groupby('Category')['Units Sold'].sum().sort_values(ascending=False)
category_sales.plot(kind='bar', color=['#3498db', '#e67e22', '#2ecc71', '#e74c3c'])
plt.title('Total Units Sold by Product Category', fontsize=14)
plt.xlabel('Product Category', fontsize=12)
plt.ylabel('Total Units Sold', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()


# Line Graph
plt.figure(figsize=(12, 6))
df_sorted = df.sort_values('Date')
plt.plot(df_sorted['Date'], df_sorted['Revenue (USD)'], marker='o', linestyle='-', color='#8e44ad', markersize=4)
plt.title('Daily Revenue Trend (USD)', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Revenue (USD)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()


# Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Units Sold', y='Revenue (USD)', hue='Category', s=100)
plt.title('Correlation: Units Sold vs. Revenue', fontsize=14)
plt.xlabel('Units Sold', fontsize=12)
plt.ylabel('Revenue (USD)', fontsize=12)
plt.legend(title='Category')
plt.grid(True, linestyle=':', alpha=0.6)
plt.tight_layout()
```
