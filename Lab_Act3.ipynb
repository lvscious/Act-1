```python
import pandas as pd

df = pd.read_csv("/content/Student_performance_factors.csv")
df.head()
```

```python
df.info()
```

```python
mean_val = df['Exam_Score'].mean()
print(f"Mean: {mean_val}")
```

```python
median_val = df['Exam_Score'].median()
print(f"Median: {median_val}")
```

```python
mode_val = df['Exam_Score'].mode()
print(f"Mode: {mode_val}")
```

```python
std_val = df['Exam_Score'].std()
print(f"Standard Deviation: {std_val}")
```

```python
min_val = df['Exam_Score'].min()
print(f"Minimum: {min_val}")
```

```python
max_val = df['Exam_Score'].max()
print(f"Maximum: {max_val}")
```

```python
df["Exam_Score"].value_counts()
df["Exam_Score"].value_counts(normalize=True)*100
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df["Exam_Score"], bins=10)
plt.title("Histogram")
plt.show()
```

```python
sns.boxplot(x=df['Attendance'])
plt.title("Box Plot")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.kdeplot(df["Hours_Studied"])
plt.title("KDE Plot of Variable")
plt.xlabel("Variable")
plt.ylabel("Density")
plt.show()
```

```python
sns.countplot(x="Gender", data=df)
plt.title("Bar Chart")
plt.show()
```

```python
df["Motivation_Level"].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Pie Chart")
plt.ylabel("")
plt.show()
```
